```python
# Import numpy library
import numpy as np
```

```python
# Set seed
np.random.seed(15)

# Initialize empty list
hw2 = []

# Generate 50 random numbers from uniform distribution with a min of 4 and a max of 50
hw2 = np.random.uniform(low = 4, high = 50, size = 50)

# Replace indices 4, 12, 22, and 27 with missing values
indices = [4, 12, 22, 27]
for index in indices:
    hw2[index] = -1
print(hw2)
print(hw2[4]==-1)
```

### Problem 1

```python
# For loop to remove missing values
inter_list = []
for value in hw2:
    if value != -1: 
        inter_list.append(value)
print(inter_list)

# Select all numbers between 14 and 38 inclusive
prob1 = []
for value in inter_list:
    if value > 13 and value < 39: 
        prob1.append(value)
print(prob1)
```

### Problem 2

```python
# Multiply each number in prob1 by 3
times3 = np.array(prob1)*3
print(times3)

# An alternative method (list comprehension) to multiple each number in prob1 by 3
times3_alt = [num*3 for num in prob1]
print(times3_alt)

# Add 10 to each number in times3
plus10 = times3 + 10
print(plus10)

# Add 10 to each number in times3 using list comprehension
plus10_alt = [num+10 for num in times3_alt]
print(plus10_alt)
```
