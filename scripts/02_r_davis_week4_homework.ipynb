```python
# Import libraries
import pandas as pd
from dplython import (DplyFrame, X, diamonds, select, sift,
  sample_n, sample_frac, head, arrange, mutate, group_by,
  summarize, DelayFunction)
# dplython is a python equivalent of dplyr which seems to have similar functionality 
```

```python
# Read in surveys data
surveys_original = DplyFrame(pd.read_csv("../data/portal_data_joined.csv"))
```

```python
# Subset surveys to keep rows with weight betweem 30 and 60
surveys = surveys_original >> sift((X.weight > 30.) & (X.weight < 60.) | (X.weight.isna() == True))
print(surveys["weight"].max())
print(surveys["weight"].min())

# Print first six rows
surveys.head(n = 6)
```

```python
# Remove rows with NAs for weight 
surveys_narm = surveys.dropna(subset = ["weight"])

# Create new df with weights by species and sex 
biggest_critters = surveys_narm >> group_by(X.species_id, X.sex) >> summarize(max_weight = X.weight.max())
print(biggest_critters.head(n = 10))

# Arrange my new df by weight to look at the biggest and smallest combos
# Arrange sorts from smallest to largest
biggest_critters = biggest_critters >> arrange(X.max_weight)

# Smallest combos
print(biggest_critters.head(n = 10))

# Largest combos 
print(biggest_critters.tail(n = 10))
```

```python
# Explore where the NAs are concentrated
# Df including only rows with NAs for weight
surveys_nas = surveys >> sift(X.weight.isna() == True)

# View NAs by species
print("View NAs by species")
print("Tally for NAs only")
print(surveys_nas.groupby(["species_id"])["weight"].agg(["size"])) # Group by species ID and then count entries; specific column doesn't matter
print("Tally total")
print(surveys_original.groupby(["species_id"])["weight"].agg(["size"]))

# View NAs by plot type
print("View NAs by plot type")
print("Tally for NAs only")
print(surveys_nas.groupby(["plot_type"])["weight"].agg(["size"]))
print("Tally total")
print(surveys_original.groupby(["plot_type"])["weight"].agg(["size"]))

# View NAs by taxa
print("View NAs by taxa")
print("Tally for NAs only")
print(surveys_nas.groupby(["taxa"])["weight"].agg(["size"]))
print("Tally total")
print(surveys_original.groupby(["taxa"])["weight"].agg(["size"]))

# It looks to me like weight was only taken for some species, as there are only NAs under weight for many species. 
# There were lots of NAs across plot types relative to total observations.
# It looks like weights may have only be taken for some rodent species, and no other taxa. 
```

```python
# I've already removed NAs for surveys_narm above
# Calculate the average weight of each species and sex combination
average_critters = surveys_narm >> group_by(X.species_id, X.sex) >> summarize(avg_weight = X.weight.mean())
print(average_critters.head)

# Merge my average weights with the full surveys dataframe 
surveys_avg_weight = surveys_narm.merge(average_critters, left_on = ["species_id", "sex"], right_on = ["species_id", "sex"])

# Select only desired columns
surveys_avg_weight = surveys_avg_weight >> select(X.species_id, X.sex, X.weight, X.avg_weight)
print(surveys_avg_weight.head)
```

```python
# Add a column specifying whether an animal's weight is above average (True) or not (False) for its species and sex
surveys_avg_weight = surveys_avg_weight >> mutate(above_average = (X.weight > X.avg_weight))
print(surveys_avg_weight.head)
```
